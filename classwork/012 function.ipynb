```python
def addition(*args): 
    # supports variable / any no. of positional arguments
    sum = 0
    for i in args:
        sum += i
        
    return sum
```

```python
addition(7, 8, 9, 10, 11, 12,14)
```

```python
addition(a=1,b=2,c=3)
```

```python
def addition(**kwargs): 
    # supports variable / any no. of keyword arguments
    print(kwargs, type(kwargs))
```

```python
addition(a=1,b=2,c=3)
```

```python
def addition(**kwargs): 
    # supports variable / any no. of keyword arguments
    sum = 0
    for i in kwargs.values():
        sum += i
        
    return sum
```

```python
addition(a=1,b=2,c=3)
```

```python
addition(1,2,3,a=22, b=33, c =77)
```

```python
def addition(*args, **kwargs): 
    print(args, type(args))
    print(kwargs, type(kwargs))
```

```python
addition(1,2,3,a=22, b=33, c =77)
```

```python
def addition(*args, **kwargs): 
    sum = 0
    
    for i in args:
        sum += i
        
    for i in kwargs.values():
        sum += i
        
    return sum
```

```python
addition(1,2,3,a=22, b=33, c =77)
```

```python
def addition(*xyz, **abc): 
    sum = 0
    
    for i in xyz:
        sum += i
        
    for i in abc.values():
        sum += i
        
    return sum
```

```python
addition(1,2,3,a=22, b=33, c =77)
```

```python
factorial(5) = 5 * 4 * 3 * 2 * 1
factorial(10) = 10 * 9 * 8 ... 1
```

```python
def factorial(n):
    fact = 1
    for i in range(n, 0, -1):
        fact *= i
    return fact
```

```python
factorial(5)
```

```python
fact(5) = 5 * 4 * 3 * 2 * 1
fact(5) = 5 * fact(4)
fact(5) = 5 * 4 * fact(3)
fact(5) = 5 * 4 * 3 * fact(2)
fact(5) = 5 * 4 * 3 * 2 * fact(1)
fact(5) = 5 * 4 * 3 * 2 * 1
```
